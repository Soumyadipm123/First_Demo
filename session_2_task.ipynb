### `Problem 1`: Write a program that will give you in hand monthly salary after deduction on CTC - HRA(10%), DA(5%), PF(3%) and taxes deduction as below:

> Salary(Lakhs) : Tax(%)

*   Below 5 : 0%
*   5-10 : 10%
*   10-20 : 20%
*   aboove 20 : 30%

````python
ctc = float(input("Enter your CTC in Lakh: "))

# Deductions
hra = ctc * 0.10
da = ctc * 0.05
pf = ctc * 0.03
n = ctc - (hra + da + pf)

# Tax calculation
if n < 5:
  final = n
elif n < 10:
    final = n * 0.90  # 10% tax deduction
elif n < 20:
    final = n * 0.80  # 20% tax deduction
else:
    final = n * 0.70  # 30% tax deduction

print(f"If your CTC is {ctc} Lakh, then your in-hand salary is {final:.2f} Lakh")
print(f"your monthly salary is {final/12}thousand only")
````

```bash
30
```


###`Problem 11`: A robot moves in a plane starting from the original point (0,0). The robot can move toward UP, DOWN, LEFT and RIGHT with a given steps.
The trace of robot movement is shown as the following:
```
UP 5
DOWN 3
LEFT 3
RIGHT 2
!
```
> The numbers after the direction are steps.

> `!` means robot stop there.

**Please write a program to compute the distance from current position after a sequence of movement and original point.**

*If the distance is a float, then just print the nearest integer.*

Example:

`Input`:
```
UP 5
DOWN 3
LEFT 3
RIGHT 2
!
```
`Output`:
```
2
```

````python
#here input is given in a string. so we have to first split and make it iterable list
moves = input("enter the robot trace : ").split()
# intialise the center point with x = 0 and y = 0
x = 0
y = 0

for i in range(0, len(moves), 2):
  directions = moves[i]

  if directions == "!":
    break

  steps = int(moves[i + 1])

  if directions == "up":
    y = y + steps

  elif directions == "down":
    y = y - steps

  elif directions == "right":
    x = x + steps
  else:
    x = x - steps

dist = (x**2 + y**2)**0.5
dist = int(dist)

print(f"distance from center to final point is {dist} unit")


````

###`Problem 12`:Write a program to print whether a given number is a prime number or not

````python
n = int(input("enter the number: "))

if n <= 1:
  print(f"{n} is not a prime number!!")
else:
  n_prime = True
  for i in range(2, int(n**0.5)+1):    # range should start from 2
    if n % i == 0:
      n_prime = False
  if n_prime:
    print(f"{n} is a prime number!!")
  else:
    print(f"{n} is not prime!!")
````

###`Problem 13`:Print all the Armstrong numbers in a given range.
Range will be provided by the user<br>
Armstrong number is a number that is equal to the sum of cubes of its digits. For example 0, 1, 153, 370, 371 and 407 are the Armstrong numbers.

````python
a,b = map(int,input("enter the range: ").split())
for i in range(a, b+1):
  # here to get each digits we will do string conversion
  s = str(i)
  sum = 0
  for j in range(0, len(s)):
    sum = sum + int(s[j])**3
  if sum == i:
    print(i, end = " ")
````
